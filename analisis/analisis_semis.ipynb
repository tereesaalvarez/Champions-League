## Modelo

#### Pasos que se seguirán en primera fase
 1. Integrar el rendimiento histórico con las estadísticas normalizadas
 2. Agregar las estadísticas de jugadores agregadas a nivel de equipo
 3. Crear variables dummy para los nombres de los equipos.
 4. Preparar el dataset final para el modelado

#### 1. Itegración de los datasets de rendimiento y estadísticas normalizadas


```python
# Unimos los datos de rendimiento con las estadísticas normalizadas basándonos en los nombres de los equipos.
# Primero, necesitamos asegurarnos de que los nombres de los equipos coincidan entre los dos datasets.

# Normalizamos los nombres de los equipos en ambos datasets para asegurar la coherencia.
df_rendimiento_semis['Equipos'] = df_rendimiento_semis['Equipos'].str.strip()
df_estadisticas_normalizadas['HOME_TEAM'] = df_estadisticas_normalizadas['HOME_TEAM'].str.strip()

# Realizamos la unión de los datos
combined_data = pd.merge(df_rendimiento_semis, df_estadisticas_normalizadas, left_on='Equipos', right_on='HOME_TEAM', how='inner')

# Eliminamos cualquier columna duplicada y revisamos los primeros registros del dataset combinado.
combined_data.drop(columns=['HOME_TEAM'], inplace=True)  # Eliminar columna duplicada
combined_data.head()
```

#### 2. Agregar estadísticas de los jugadores de cada equipo.

```python

# Función para calcular estadísticas agregadas
def calculate_aggregated_stats(df):
    return pd.Series({
        'Total_Goles': df['Goles'].sum(),
        'Total_Asistencias': df['Asitencias'].sum(),
        'Promedio_Rating': df['Rating'].mean(),
        'Total_Minutos': df['Minutos'].sum(),
        'Total_Amarillas': df['Amarillas'].sum(),
        'Total_Rojas': df['Roja'].sum()
    })

# Aplicamos la función a cada dataframe de jugadores y asignamos a cada equipo
bayern_stats = calculate_aggregated_stats(datosbayern)
madrid_stats = calculate_aggregated_stats(datosmadrid)
psg_stats = calculate_aggregated_stats(datospsg)
dortmund_stats = calculate_aggregated_stats(datosdortmund)

# Agregamos estas estadísticas al dataframe combinado según el equipo
team_stats = {
    "Bayern München": bayern_stats,
    "Real Madrid": madrid_stats,
    "Paris Saint-Germain": psg_stats,
    "Borussia Dortmund": dortmund_stats
}

for team, stats in team_stats.items():
    combined_data.loc[combined_data['Equipos'] == team, stats.index] = stats.values

combined_data.to_csv('datos_semis/combined_data.csv', index=False)

combined_data.head()
```

#### 3 y 4. Preparar el dataset para el modelado

##### Pasos:
 1. Crear variables dummy para los equipos
 2. Normalización o estandarización de datos
 3. División del dataset

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Creamos variables dummy para los nombres de los equipos
combined_data_dummies = pd.get_dummies(combined_data, columns=['Equipos'])

# Seleccionamos las columnas numéricas para la normalización
numeric_columns = combined_data_dummies.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
combined_data_dummies[numeric_columns] = scaler.fit_transform(combined_data_dummies[numeric_columns])

# Dividimos el dataset en conjuntos de entrenamiento y prueba (suponiendo que más adelante definiremos 'y')
X_train, X_test = train_test_split(combined_data_dummies, test_size=0.2, random_state=42)

# Mostramos las primeras filas del dataset de entrenamiento para verificar las transformaciones
X_train.head(5)
```

```python
# Aplicamos nuevamente la creación de variables dummy y normalización a todo el dataset, sin dividirlo esta vez.
combined_data_dummies_full = pd.get_dummies(combined_data, columns=['Equipos'])

# Normalizamos las columnas numéricas
combined_data_dummies_full[numeric_columns] = scaler.fit_transform(combined_data_dummies_full[numeric_columns])

# Mostramos las filas del dataset transformado
combined_data_dummies_full
```

## Modelo de Regresión Logística

### Paso 1: Simular los resultados de los partidos

Voy a simular los resultados basados en el 'Promedio de Rating' para los enfrentamientos Madrid vs Bayern y Psg vs Dortmund. Después simularé el ganador de la final basado en estos resultados simulados.

```python
# Supongo que la columna 'Promedio_Rating' puede ayudarnos a determinar el ganador.
# Simulo los resultados usando el promedio de rating como criterio.
# Extraigo los datos relevantes para los enfrentamientos.

# Madrid vs Bayern
madrid_rating = combined_data_dummies_full.loc[combined_data_dummies_full['Equipos_Real Madrid'] == 1, 'Promedio_Rating'].values[0]
bayern_rating = combined_data_dummies_full.loc[combined_data_dummies_full['Equipos_Bayern München'] == 1, 'Promedio_Rating'].values[0]

# PSG vs Dortmund
psg_rating = combined_data_dummies_full.loc[combined_data_dummies_full['Equipos_Paris Saint-Germain'] == 1, 'Promedio_Rating'].values[0]
dortmund_rating = combined_data_dummies_full.loc[combined_data_dummies_full['Equipos_Borussia Dortmund'] == 1, 'Promedio_Rating'].values[0]

# Simulación de resultados
# 1 si el equipo de casa gana, 0 si el equipo visitante gana
results = {
    'madrid_vs_bayern': 1 if madrid_rating > bayern_rating else 0,
    'psg_vs_dortmund': 1 if psg_rating > dortmund_rating else 0
}

# Determinar los equipos en la final
final_teams = {
    'final_team_1': 'Real Madrid' if results['madrid_vs_bayern'] == 1 else 'Bayern München',
    'final_team_2': 'Paris Saint-Germain' if results['psg_vs_dortmund'] == 1 else 'Borussia Dortmund'
}

# Rating de los equipos finalistas
final_team_1_rating = combined_data_dummies_full.loc[combined_data_dummies_full[f'Equipos_{final_teams["final_team_1"]}'] == 1, 'Promedio_Rating'].values[0]
final_team_2_rating = combined_data_dummies_full.loc[combined_data_dummies_full[f'Equipos_{final_teams["final_team_2"]}'] == 1, 'Promedio_Rating'].values[0]

# Resultado de la final
results['final'] = 1 if final_team_1_rating > final_team_2_rating else 0

results
```

Los resultados simulados según el 'Promedio de rating' son:
 - Madrid vs Bayern: Madrid gana (1 indica que el equipo local, Madrid, gana).
 - PSG vs Dortmund: Dortmund gana (0 indica que el equipo visitante, Dortmund, gana).
 - Final (Real Madrid vs Dortmund): Real Madrid gana.

### Paso 2: Entrenamiento del Modelo

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Configuramos los datos de entrenamiento para los enfrentamientos
# Características (quitamos las columnas de los equipos para evitar fugas de datos)
X = combined_data_dummies_full.drop(columns=[col for col in combined_data_dummies_full.columns if 'Equipos_' in col])

# Resultados de los enfrentamientos
y_madrid_vs_bayern = results['madrid_vs_bayern']
y_psg_vs_dortmund = results['psg_vs_dortmund']
y_final = results['final']

# Modelo para Madrid vs Bayern
model_madrid_vs_bayern = LogisticRegression()
model_madrid_vs_bayern.fit(X, [y_madrid_vs_bayern, y_madrid_vs_bayern, 1-y_madrid_vs_bayern, 1-y_madrid_vs_bayern])  # Duplicamos entradas para ajustar

# Modelo para PSG vs Dortmund
model_psg_vs_dortmund = LogisticRegression()
model_psg_vs_dortmund.fit(X, [y_psg_vs_dortmund, y_psg_vs_dortmund, 1-y_psg_vs_dortmund, 1-y_psg_vs_dortmund])  # Duplicamos entradas para ajustar

# Modelo para la final
model_final = LogisticRegression()
model_final.fit(X, [y_final, y_final, 1-y_final, 1-y_final])  # Duplicamos entradas para ajustar

# Predicciones (usamos los mismos datos por simplicidad)
pred_madrid_vs_bayern = model_madrid_vs_bayern.predict(X)
pred_psg_vs_dortmund = model_psg_vs_dortmund.predict(X)
pred_final = model_final.predict(X)

# Evaluamos la precisión y la matriz de confusión para uno de los modelos (ejemplo con Madrid vs Bayern)
accuracy_madrid_vs_bayern = accuracy_score([y_madrid_vs_bayern, y_madrid_vs_bayern, 1-y_madrid_vs_bayern, 1-y_madrid_vs_bayern], pred_madrid_vs_bayern)
conf_matrix_madrid_vs_bayern = confusion_matrix([y_madrid_vs_bayern, y_madrid_vs_bayern, 1-y_madrid_vs_bayern, 1-y_madrid_vs_bayern], pred_madrid_vs_bayern)

print(accuracy_madrid_vs_bayern)
print(conf_matrix_madrid_vs_bayern)
```
